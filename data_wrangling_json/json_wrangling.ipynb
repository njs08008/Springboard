## Find the 10 Most Common Project Themes

Let's first further inspect the `mjtheme_namecode` feature we were told to use.

```python
json_df['mjtheme_namecode'][0]
```

We see that the first project actually had two themes.  Indeed, most projects have multiple themes.
Further, the name of the second theme is missing (as the instructions warned).
This information is nested in dictionaries that are not easily accessible as we currently have the data stored.
Instead, read the data as a list of dictionaries utilizing its natural structure.

```python
# Read the json file into a list of dictionaries:
with open('data/world_bank_projects.json', 'r') as json_file:
    json_dict = json.load(json_file)
```

```python
# Create a dataframe of values for the nested element 'mjtheme_namecode':
themes_df=json_normalize(json_dict,'mjtheme_namecode')
themes_df.head(10)
```

Since `name` has some missing entries, instead compute the most common project themes using `code`.

```python
# Compute a list of the top 10 project themes by code number:
themes_df['code'].value_counts().head(10)
```

This information is not very enlightening without knowing what each `code` value represents, so let's go back and redo this calculation using `name`.

## Fill in Missing `name` Values
In order to perform the calculation using `name`, there cannot be any missing values.  To fix this problem, three steps will be taken:
1. Sort the data by `code` and then `name` so all like missing values are together.
2. Change all missing values to `NaN`.
3. Change all missing values to the correct value using surrounding data.

```python
# Sort project themes by code and then by name:
themes_df=themes_df.sort_values(['code','name'])
themes_df.head(10)
```

Notice that this results in all projects of a fixed `code` with missing `name` appearing before the projects of the same `code` whose `name` entry is not missing. 

```python
# Change all blank project theme names to NaN
themes_df.name[themes_df['name']=='']=np.nan
```

Since the missing `name` values appear before the actual values, we can backfill them.

```python
# backfill the NaNs created to give proper names:
themes_df=themes_df.fillna(method='bfill')
```

```python
# Recompute the most common project themes, this time by their descriptive names.
themes_df['name'].value_counts()
```
